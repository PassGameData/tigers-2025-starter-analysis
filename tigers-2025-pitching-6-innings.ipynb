```python
from pybaseball import statcast
import pandas as pd

statcast_data = statcast(start_dt="2025-03-01", end_dt="2025-10-01")
statcast_data.shape
```

```python
statcast_data.head()
```

```python
statcast_data.shape
```

```python
tigers_pitches = statcast_data[
    (statcast_data['home_team'] == 'DET')|
    (statcast_data['away_team'] == 'DET')
]

tigers_pitches.shape
```

```python
non_out_events = [
    'single',
    'double',
    'triple',
    'home_run',
    'walk',
    'hit_by_pitch',
    'intent_walk',
    'error',
    'field_error'
]

tigers_pitches['is_out'] = (
    tigers_pitches['events'].notna() &
    (~tigers_pitches['events'].isin(non_out_events))
)

tigers_pitches['outs'] = tigers_pitches['is_out'].astype(int)
```

```python
pitching_game_level = (
    tigers_pitches
    .groupby(['game_pk', 'pitcher', 'player_name'], as_index=False)
    .agg({'outs': 'sum'})
)

pitching_game_level.head()
```

```python
pitching_game_level.shape
```

```python
starter_per_game = (
    pitching_game_level
    .sort_values(['game_pk', 'outs'], ascending=[True, False])
)

starter_per_game.head()
```

```python
starter_per_game.shape
```

```python
starter_per_game = (
    tigers_pitches
    .groupby(['game_pk', 'pitcher', 'player_name'], as_index=False)
    .agg(
        outs=('events', lambda x: sum(
            ~x.isin([
                'single','double','triple','home_run',
                'walk','hit_by_pitch','intent_walk','error','field_error'
            ]) & x.notna()
        )),
        first_inning=('inning', 'min')
    )
)
```

```python
starter_per_game = (
    starter_per_game
    .sort_values(['game_pk', 'first_inning', 'outs'], ascending=[True, True, False])
    .drop_duplicates(subset=['game_pk'], keep='first')
)

starter_per_game.shape
```

```python
starter_per_game[starter_per_game['outs'] >= 18]
```

```python
six_plus_starters = starter_per_game[starter_per_game['outs'] >= 18]

six_plus_starters.shape
```

```python
statcast_games = tigers_pitches[['game_pk', 'game_date']].drop_duplicates()
```

```python
six_plus_starters = six_plus_starters.merge(
    statcast_games,
    on='game_pk',
    how='left'
)
```

```python
six_plus_starters['game_date'] = pd.to_datetime(six_plus_starters['game_date'])
```

```python
final_df = six_plus_starters.merge(
    tigers_schedule[['Date', 'W/L']],
    left_on='game_date',
    right_on='Date',
    how='left'
)
```

```python
wins = final_df[final_df['W/L'] == 'W'].shape[0]
losses = final_df[final_df['W/L'] == 'L'].shape[0]

print("Record when starter went 6+ innings:")
print(f"{wins} - {losses}")
```

```python
win_pct = wins / (wins + losses)

print(f"Win % when starter went 6+ innings: {win_pct:.3f}")
```

```python
summary = pd.DataFrame({
    "Condition": ["Starter ≥ 6 IP"],
    "Wins": [wins],
    "Losses": [losses],
    "Win_Percentage": [win_pct]
})

summary
```
